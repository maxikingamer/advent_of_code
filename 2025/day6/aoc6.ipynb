```python
lines = []

with open("input6.txt") as file:
    for line_n, line in enumerate(file):
        if line_n <= 3:
            line = line.replace(" ", ",")
            line = line.split(",")
            line = [int(x) for x in line if len(x)>0]
            lines.append(line)
        else:
            line = line.replace(" ", ",")
            line = line.split(",")
            line = [x for x in line if len(x)>0]
            operations = line
```

```python
import numpy as np

numbers = np.matrix(lines).T
```

```python
complete_sum = 0
for number_row, operation in zip(numbers, operations):
    number_row = np.squeeze(list(number_row))
    print(number_row)
    print(operation)
    if operation == "+":
        complete_sum += sum(number_row)
    elif operation == "*":
        tmp = number_row[0]
        for number in number_row[1:]:
            tmp *= number
        complete_sum += tmp
```

```python

all_numbers = []
operations = []
numbers = []

whatever = []

for i in range(0, 3724):
    with open("input6.txt") as file:
        tmp = []
        for line in file:
            tmp.append(line[i])
        whatever.append(tmp)



for element in whatever:
    if all([x==" " for x in element]):
        all_numbers.append(numbers)
        numbers = []
        continue
    number = []
    for character in element:
        if character.isdigit():
            number.append(character)
        elif character in ["*", "+"]:
            operations.append(character)

    numbers.append("".join(number))
            

        
```
