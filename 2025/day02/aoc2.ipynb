```python
with open("input2.txt") as file:
    for line in file:
        input_line = line
ranges = input_line.split(",")
print(ranges)
```

```python
invalid_sum = 0
for one_range in ranges:
    start, end = one_range.split("-")
    for i in range(int(start), int(end)+1):
        string_i = str(i)
        if len(string_i) % 2 == 0:
            first_half = int(string_i[:int(len(string_i)/2)])
            second_half = int(string_i[int(len(string_i)/2):])
            if first_half == second_half:
                invalid_sum += i
```

```python
print(invalid_sum)
```

```python
invalid_sum = 0
for one_range in ranges:
    start, end = one_range.split("-")
    for i in range(int(start), int(end)+1):
        string_i = str(i)
        divisors = []
        for j in range(1, len(string_i)+1):
            if len(string_i) % j == 0:
                divisors.append(j)

        tmp = []
        for divisor in divisors:
            k = 0
            parts = []
            while k < len(string_i):
                parts.append(string_i[k:k+divisor])
                k += divisor

            if len(parts) == 1:
                continue
            elif len(set(parts)) == 1:
                tmp.append(i)
        invalid_sum += sum(set(tmp))
```
