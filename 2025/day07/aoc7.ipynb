```python
grid = []
with open("input7.txt") as file:
    for line in file:
        # grid.append(list(line[:141]))
        grid.append(list(line[:15]))

grid = np.matrix(grid)

start = np.where(grid == "S")
sources = [(start[0][0], start[1][0])]
split_count = 0

n_timelines = 0

def parse_down(grid, sources):
    global n_timelines

    if n_timelines > 100000000:
        return None

    while len(sources) > 0:
        for source in sources:
            if source[0] < grid.shape[1] - 1:
                if grid[source[0]+1,source[1]] == ".":
                    grid[source[0]+1,source[1]] = "|"
                    sources.append((source[0]+1,source[1]))

                elif grid[source[0]+1,source[1]] == "^":
                    if source[1] > 0 & source[1] < grid.shape[0]-1:
                        grid[source[0]+1, source[1]-1] = "|"
                        grid[source[0]+1, source[1]+1] = "|"
                        # print("Splitting Time at ", source)
                        if source in sources:
                            new_sources_left = sources.copy()
                            new_sources_left.append((source[0]+1, source[1]-1))
                            new_sources_left.remove(source)

                            new_sources_right = sources.copy()
                            new_sources_right.append((source[0]+1, source[1]+1))
                            new_sources_right.remove(source)

                        return parse_down(grid.copy(), new_sources_left), parse_down(grid.copy(), new_sources_right)
                    
                    elif source[1] > 0:
                        grid[source[0]+1, source[1]-1] = "|"
                        # print("Splitting Time at ", source)
                        if source in sources:
                            new_sources_left = sources.copy()
                            new_sources_left.append((source[0]+1, source[1]-1))
                            new_sources_left.remove(source)
                        
                        return parse_down(grid.copy(), new_sources_left)
                    
                    elif source[1] < grid.shape[0]-1:
                            grid[source[0]+1, source[1]+1] = "|"
                            # print("Splitting Time at ", source)
                            if source in sources:
                                new_sources_right = sources.copy()
                                new_sources_right.append((source[0]+1, source[1]+1))
                                new_sources_right.remove(source)
                            return parse_down(grid.copy(), new_sources_right)
            else:
                if grid[source[0]+1,source[1]] == ".":
                    grid[source[0]+1,source[1]] = "|"

            sources.remove(source)

    n_timelines += 1
    
    if n_timelines % 1 == 0:
        print(n_timelines)
        np.savetxt(f'timelines/timeline_{n_timelines}.out', grid, delimiter=',', header='string', comments='', fmt='%s') 

    return None

parse_down(grid, [start])
```

```python
import numpy as np

grid_list = []
with open("input7.txt") as file:
    for line in file:
        grid_list.append(list(line.strip()[:141]))
grid = np.array(grid_list)

start_pos = np.where(grid == "S")
start_node = (start_pos[0][0], start_pos[1][0])

memo = {}

def count_unique_paths(r, c):
    if (r, c) in memo:
        return memo[(r, c)]
    
    if r >= grid.shape[0] - 1:
        return 1 

    total_paths = 0
    next_r = r + 1

    if grid[next_r, c] == "." or grid[next_r, c] == "|":
        total_paths += count_unique_paths(next_r, c)
        
    elif grid[next_r, c] == "^":
        if c > 0:
            total_paths += count_unique_paths(next_r, c - 1)
        if c < grid.shape[1] - 1:
            total_paths += count_unique_paths(next_r, c + 1)
            
    memo[(r, c)] = total_paths
    return total_paths

unique_paths = count_unique_paths(start_node[0], start_node[1])
print(f"Total Unique Paths: {unique_paths}")
```
