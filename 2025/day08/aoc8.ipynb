```python
from itertools import combinations  
import math
import pandas as pd
```

```python
def get_distance(box1, box2):
    return math.sqrt((box1[0] - box2[0])**2 + (box1[1] - box2[1])**2 + (box1[2] - box2[2])**2)
```

```python
all_boxes = []

with open("input8.txt") as file:
    for line_n, line in enumerate(file):
        line = line.replace("\n", "")
        line = [int(x) for x in line.split(",")]
        line.append(line_n)
        line.append(None)
        all_boxes.append(line)

used_connections = []
check_connections = all_boxes.copy()

for i in range(0,1000):
    print(i)
    min_distance = math.inf
    for b1, b2 in list(combinations(all_boxes, 2)):
        if (b1, b2) in used_connections:
            continue

        distance = get_distance(b1, b2)

        if distance < min_distance:
            min_distance = distance
            min_pair = (b1.copy(),b2.copy())
            min_ids = (b1[3], b2[3])

    all_boxes[min_ids[0]][4] = i
    all_boxes[min_ids[1]][4] = i

    used_connections.append((all_boxes[min_ids[0]], all_boxes[min_ids[1]]))


    for j, box in enumerate(all_boxes):
        if min_pair[0][4] != None and min_pair[1][4] != None:
            if box[4] == min_pair[0][4] or box[4] == min_pair[1][4]:
                all_boxes[j][4] = i
        elif min_pair[0][4] != None:
            if box[4] == min_pair[0][4]:
                all_boxes[j][4] = i
        elif min_pair[1][4] != None:
            if box[4] == min_pair[1][4]:
                all_boxes[j][4] = i

    
```
