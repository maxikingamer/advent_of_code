```python
with open("input12.txt") as file:
    content = file.read()

blocks = content.split("\n\n")
boxes = [s.replace("\n", "")[2:] for s in blocks[:-1]]
grids = blocks[-1].split("\n")
```

```python
def count_spaces(box, n):
    spaces_per_box = box.count('#')
    return spaces_per_box * n
```

```python
sure_fits_counter = 0
sure_nofits_counter = 0
for grid in grids:
    dimensions, presents = grid.split(":")
    space = [int(x) for x in dimensions.split("x")]
    presents = [int(x) for x in presents.split(" ")[1:]]

    print(grid)

    min_spaces_needed = 0
    for shape_index, present in enumerate(presents):
        min_spaces_needed += count_spaces(boxes[shape_index], int(present))

    if space[0] * space[1] >= sum(presents) * 9:
        sure_fits_counter += 1

    if min_spaces_needed > space[0] * space[1]:
        sure_nofits_counter += 1
```
