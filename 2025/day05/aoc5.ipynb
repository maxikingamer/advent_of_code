```python
ranges = []
fresh_items = 0
unique_ranges = []
with open("input5.txt") as file:
    is_range = True
    for line_n, line in enumerate(file):
        if line == "\n":
            is_range = False
            continue

        line = line.replace('\n', '')

        if is_range:
            start, end = line.split("-")
            current_range = (int(start), int(end))
            ranges.append(current_range)
                   
        else:
            item_id = int(line)
            for range in ranges:
                start, end = range
                if start <= item_id <= end:
                    fresh_items += 1
                    break
```

```python
sorted_ranges = sorted(ranges)
unique_ranges = []
for sorted_range in sorted_ranges:
    if len(unique_ranges) == 0:
        unique_ranges.append(sorted_range)
    else:
        if sorted_range[1] > unique_ranges[-1][1]:
            if sorted_range[0] <= unique_ranges[-1][1]:
                unique_ranges.append((unique_ranges[-1][1]+1, sorted_range[1]))
            else:
                unique_ranges.append(sorted_range)
            

   
```
