```python
wiring_dictionary = {}
with open("input11.txt") as file:
    for line_n, line in enumerate(file):
        clean_line = line.replace("\n", "").replace(":", "")
        clean_line = clean_line.split(" ")
        wiring_dictionary[clean_line[0]] = clean_line[1:]

counter = 0

def traverse_cables(current_cable):
    global counter
    if current_cable == "out":
        counter += 1
        return None
    
    neighbours = wiring_dictionary[current_cable]
    for neighbour in neighbours:
        traverse_cables(neighbour)

traverse_cables("you")
```

```python
def solve_wiring(wiring_dictionary, start_node="svr", end_node="out"):
    memo = {}

    def traverse(current_node, has_fft, has_dac, visited):
        if current_node == "fft":
            has_fft = True
        elif current_node == "dac":
            has_dac = True

        if current_node == end_node:
            return 1 if (has_fft and has_dac) else 0

        state = (current_node, has_fft, has_dac)
        if state in memo:
            return memo[state]

        total_paths = 0
        neighbours = wiring_dictionary[current_node]
        
        new_visited = visited | {current_node}

        for neighbour in neighbours:
            if neighbour not in visited:
                total_paths += traverse(neighbour, has_fft, has_dac, new_visited)

        memo[state] = total_paths
        return total_paths

    return traverse(start_node, False, False, set())


total_good_paths = solve_wiring(wiring_dictionary)
print(f"Total valid paths found: {total_good_paths}")
```
