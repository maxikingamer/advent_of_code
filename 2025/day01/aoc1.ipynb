```python
def turn_left(start, n):
    if start != 0:
        n_zeros = int((start <= (n % 100))) + (n // 100)
    else:
        n_zeros = n // 100
    return (start - n) % 100, n_zeros

def turn_right(start, n):
    if start != 0:
        n_zeros = int(((n % 100) + start) >= 100) + (n // 100)
    else:
        n_zeros = n // 100
    return (start + n) % 100, n_zeros

def turn(orientation, start, n):
    if orientation == "L":
        end, n_zeros = turn_left(start, n)
    elif orientation == "R":
        end, n_zeros = turn_right(start, n)

    return end, n_zeros
```

```python
zeros = 0
n_dials_over_zero = 0
end = 50
with open("input1.txt") as file:
    for line in file:
        orientation = line[0]
        n = int(line[1:])
        print(end, line)
        end, n_zeros = turn(orientation, end, n)
        n_dials_over_zero += n_zeros
        print("(", end, ",", n_dials_over_zero, ")")
        if end == 0:
            zeros += 1
        
```
