```python
import numpy as np
import cv2
```

```python
red_tiles = []
red_x = []
red_y = []
with open("input9.txt") as file:
    for line in file:
        x, y = line.split(",")
        red_tiles.append((int(x), int(y)))
        red_x.append(int(x))
        red_y.append(int(y))

print(red_tiles)
```

```python
max_area = 0
max_nodes = []
for tile1 in red_tiles:
    for tile2 in red_tiles:
        x_check = False
        y_check = False

        if tile1[0] <= tile2[0]:
            row1 = [x for x in red_tiles if x[0] == tile1[0] and x != tile1]
            if len(row1) > 0:
                row1_max_y = max(x[1] for x in row1)
                if row1_max_y >= tile2[1]:
                    x_check = True
        else:
            row2 = [x for x in red_tiles if x[0] == tile2[0] and x != tile2]
            if len(row2) > 0:
                row2_max_y = max(x[1] for x in row2)
                if row2_max_y >= tile1[1]:
                    x_check = True

        if tile1[1] <= tile2[1]:
            row1 = [x for x in red_tiles if x[1] == tile1[1] and x != tile1]
            if len(row1) > 0:
                row1_max_x = max(x[0] for x in row1)
                if row1_max_x >= tile2[0]:
                    y_check = True
        else:
            row2 = [x for x in red_tiles if x[1] == tile2[1] and x != tile2]
            if len(row2) > 0:
                row2_max_x = max(x[0] for x in row2)
                if row2_max_x >= tile1[0]:
                    y_check = True
                
        print(x_check)
        print(y_check)
        print(tile1)
        print(tile2)
        print("----")

        if x_check and y_check:
            area = (np.abs(tile1[0] - tile2[0]) + 1) * (np.abs(tile1[1] - tile2[1]) + 1)

            print(tile1)
            print(tile2)
            print(area)
            print("----")
            
            if area > max_area:
                
                max_area = area
                max_nodes = (tile1, tile2)
```

```python
# def is_rectangle_inside(grid, p1, p2, current_max):
#     x_min, x_max = min(p1[0], p2[0]), max(p1[0], p2[0])
#     y_min, y_max = min(p1[1], p2[1]), max(p1[1], p2[1])

#     # y_min, y_max = min(p1[0], p2[0]), max(p1[0], p2[0])
#     # x_min, x_max = min(p1[1], p2[1]), max(p1[1], p2[1])
    
#     if x_min < 0 or y_min < 0 or x_max >= grid.shape[1] or y_max >= grid.shape[0]:
#         return False, None
    
#     # rect_slice = grid[y_min : y_max + 1, x_min : x_max + 1]
#     area = (x_max - x_min + 1) * (y_max - y_min + 1)

#     if area > current_max:
#         # if np.all(rect_slice == 1):
#         #     return True, area
#         if grid[x_min:x_max]

#     return False, area

def is_rectangle_inside(grid, p1, p2, current_max):
    x_min, x_max = min(p1[0], p2[0]), max(p1[0], p2[0])
    y_min, y_max = min(p1[1], p2[1]), max(p1[1], p2[1])

    width = x_max - x_min + 1
    height = y_max - y_min + 1
    area = width * height

    if area <= current_max:
        return False, area

    target_area = grid[y_min : y_max + 1, x_min : x_max + 1]
    
    if np.all(target_area == 1):
        return True, area
    
    return False, area


def is_rectangle_all_ones_fast(integral_grid, p1, p2, current_max):
    x1, x2 = min(p1[0], p2[0]), max(p1[0], p2[0])
    y1, y2 = min(p1[1], p2[1]), max(p1[1], p2[1])
    
    width = x2 - x1 + 1
    height = y2 - y1 + 1
    area = width * height
    
    if area <= current_max:
        return False, area

    rect_sum = (integral_grid[y2+1, x2+1] - integral_grid[y1, x2+1] - 
                integral_grid[y2+1, x1] + integral_grid[y1, x1])
    
    return rect_sum == area, area

points = np.array(red_tiles)
grid = np.zeros(np.max(red_tiles, axis=0)[::-1] + [1,1], dtype=np.uint8)
filled_grid = cv2.fillPoly(grid, [points], color=1)
integral_grid = cv2.integral(filled_grid)

biggest_rect = 0
biggest_cords = []
for i, tile1 in enumerate(red_tiles):
    print(f"Iteration: {i}")
    top_contenders = red_tiles[i:]

    for j, tile2 in enumerate(top_contenders):
        maximum_area = (np.abs(tile1[0] - tile2[0]) + 1) * np.abs((tile1[1] - tile2[1] + 1))
        if maximum_area > biggest_rect:
            # if tile1[0] < tile2[0] and tile1[1] < tile2[1]:
                # check, area = is_rectangle_inside(filled_grid, tile1, tile2, biggest_rect)
                check, area = is_rectangle_all_ones_fast(integral_grid, tile1, tile2, biggest_rect)

                if check and area > biggest_rect:
                    biggest_rect = area
                    biggest_cords = [tile1, tile2]


print(biggest_rect)
print(biggest_cords)
```
