```python
import numpy as np
```

```python
def get_neighbours(col, row, grid):
    check_frame = (max(0, row-1), min(grid.shape[0], row+1), max(0, col-1), min(grid.shape[0], col+1))
    frame = grid[check_frame[0]:check_frame[1]+1, check_frame[2]:check_frame[3]+1]

    neighbours = np.sum(np.char.count(frame.flatten(), "@"))
    return neighbours, frame
```

```python
grid = []
with open("input4.txt") as file:
    for line in file:
        grid.append(list(line[:135]))

grid = np.matrix(grid)

free_rolls = 0

removed_rolls = 1
while removed_rolls > 0:
    removed_rolls = 0
    for index, element in np.ndenumerate(grid):
        if element == "@":
            neighbours, frame = get_neighbours(index[1], index[0], grid)
            if neighbours < 5:
                free_rolls += 1
                grid[index] = "."
                removed_rolls += 1

print(free_rolls)
```
